```python
import torch.nn as nn
import torch
import matplotlib.pyplot as plt
```

```python
img = plt.imread(r'../datasets/sgg-data/duck.jpg')

type(img)
```

```python
img.shape
```

```python
# （h, w, c) -> (c, h, w)
tensor = torch.tensor(img).permute(2, 0, 1).float()
print(tensor.shape)
```

```python
# 定义卷积层
conv = nn.Conv2d(3, 3, 9, 3)

y = conv(tensor)

print(y.shape)
```

```python
# 将特征图转换成图像
y = torch.clamp(y.int(), 0, 255).detach().numpy().transpose(1, 2, 0)
y.shape
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 10))
ax[0].imshow(img)
ax[1].imshow(y)
```
