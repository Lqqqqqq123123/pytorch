```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
```

```python
img = plt.imread(r'../datasets/sgg-data/duck.jpg')
type(img)
```

```python
tensor = torch.tensor(img, dtype=torch.float)
tensor.shape
```

```python
tensor = tensor.permute(2, 0, 1)
tensor.shape
```

```python
# 池化层
pool1 = nn.AvgPool2d(3, 3)
tensor = pool1(tensor)
print(tensor.shape)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(img)
tensor = tensor.permute(1, 2, 0)
ax[1].imshow(tensor.clamp(0, 255).int().detach().numpy())
```
