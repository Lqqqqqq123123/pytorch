```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

## Tanh的可视化（直接用自动微分模块求导）

```python
x = torch.linspace(-3, 3, 100, requires_grad=True)

y = torch.tanh(x)

loss = y.sum()

loss.backward()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 4))
# tanh函数可视化
x_grad = x.grad
x_plot = x.data
y_plot = y.data

ax[0].plot(x_plot, y_plot, color='r', linewidth=2, alpha=0.8)
ax[0].axhline(y=-1.0, alpha=0.2, color='gray')
ax[0].axhline(y=1.0, alpha=0.2, color='gray')
ax[0].set_title('Sigmoid Function')

# 调整坐标轴
ax[0].spines['top'].set_visible(False)
ax[0].spines['right'].set_visible(False)
ax[0].spines['left'].set_position('zero')
ax[0].spines['bottom'].set_position('zero') 

# tanh函数导数可视化  
ax[1].plot(x_plot, x_grad, color='cyan', linewidth=1, alpha=0.5)
ax[1].set_title('Sigmoid Gradient')
# 调整坐标轴
ax[1].spines['top'].set_visible(False)
ax[1].spines['right'].set_visible(False)
ax[1].spines['left'].set_position('zero')
ax[1].spines['bottom'].set_position('zero') 
plt.show()
```
